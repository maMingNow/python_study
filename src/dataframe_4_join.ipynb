```python
### 加载join需要的 主表 和 从表
df_main = pd.read_csv("/Users/maming/Downloads/test/dataframe_4_join_main.txt",sep='\t',nrows = 100,encoding = "utf-8",engine="python")
print df_main
print "-------"
df_city = pd.read_csv("/Users/maming/Downloads/test/dataframe_4_join_city.txt",sep='\t',nrows = 100,encoding = "utf-8",engine="python")
print df_city
```

```python
### 如果有相同的id,因此会省略on,但是最好还是要写on
print pd.merge(df_main,df_city,on = "id") ### print pd.merge(df_main,df_city)
## 注意 两个表都有相同的字段的时候,展示内容是类型_x 类型_y
print pd.merge(df_main,df_city,on = "id",suffixes = ["_L","_R"]) ### 更改成left 和 right
```

```python
### 多字段关联
pd.merge(df_main,df_city,on = ["id",u"类型"])
```

```python
## 如果两个表字段不一致时
## left_on  right_on 添加左边和右边表的主键。
## left_index = True/false  right_index = True/false 添加左边和右边表的索引主键
##pd.merge(df_main,df_city,left_on = ["id",u"类型"],right_on = ["id1",u"类型"])
```

```python
### 连接方式 
### 默认merge是inner join 操作,所以本来数据源有16条数据,最后只显示15条,因为最后一条没有关联上,则不显示
### how = left right outer
print pd.merge(df_main,df_city,on = "id",how = "outer",suffixes = ["_L","_R"]) ### 更改成left 和 right
```

```python
### 纵向 union all 关联
### 要求两个表的列数相同、类型相同
pd.concat([df_main,df_main])
```

```python
### 上面的demo,索引是重复的,因为默认join的时候是保留了原表的索引的
pd.concat([df_main,df_main],ignore_index = True)
```

```python
### 去掉重复数据
### 因为返回值是DataFrame,因此使用DataFrame的数据处理就可以删除重复数
print type (pd.concat([df_main,df_main],ignore_index = True)) 
pd.concat([df_main,df_main],ignore_index = True).drop_duplicates()
```
